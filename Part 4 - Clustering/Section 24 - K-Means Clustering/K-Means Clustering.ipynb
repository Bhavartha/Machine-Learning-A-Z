# K-Means Clustering<hr>

```python
# Importing required libraries

%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

plt.rcParams['figure.figsize']=(16,8)
```

```python
# Importing dataset

dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:,[3,4]].values
dataset.head()
```

```python
# Using elbow method to find optimal number of clusters
wcss=[]
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',n_init=10,max_iter=300,random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('Elbow K-Means')
plt.xlabel('No. of clusters')
plt.xticks(range(1,11))
plt.ylabel('Within Cluster sum of squares')
plt.show()
```

```python
# Applying K-Means

kmeans = KMeans(n_clusters=5,init='k-means++',n_init=10,max_iter=300,random_state=0)
y = kmeans.fit_predict(X)
y
```

```python
# Visualising

plt.style.use('fivethirtyeight')
plt.scatter(X[y==0,0],X[y==0,1],s=100,c='red',label='C1')
plt.scatter(X[y==1,0],X[y==1,1],s=100,c='green',label='C2')
plt.scatter(X[y==2,0],X[y==2,1],s=100,c='blue',label='C3')
plt.scatter(X[y==3,0],X[y==3,1],s=100,c='cyan',label='C4')
plt.scatter(X[y==4,0],X[y==4,1],s=100,c='orange',label='C5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='black',label='Centroid',marker='x')

plt.xlabel('Annual Income k$')
plt.ylabel('Spending Score')

plt.legend(shadow=True)
plt.show()
```
